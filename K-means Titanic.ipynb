```python

# 1 Importar libreras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
import seaborn as sns

# 2 Cargar datos
file_path = "/Users/andres/Documents/Codigo/titanic.csv" 
df = pd.read_csv(file_path)

# 3 Seleccionar columnas relevantes
cols = ["Age", "Fare", "Pclass", "Survived"]
df = df[cols]

# 4 Limpiar datos (eliminar filas con valores nulos)
df = df.dropna()

# 5 Escalar variables
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# 6 Aplicar K-means
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(scaled_data)

# 7 Visualizar resultados
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df, 
    x="Age", 
    y="Fare", 
    hue="Cluster", 
    palette="viridis"
)
plt.title("Clusters generados por K-Means (Titanic)")
plt.show()



# 9 Mostrar promedios por cluster
cluster_summary = df.groupby("Cluster")[cols].mean()
print("\nPromedios de cada cluster:")
print(cluster_summary)

#   Grfico comparando 'Survived' y 'Cluster'
plt.figure(figsize=(7,5))
sns.countplot(x="Cluster", hue="Survived", data=df, palette="Set2")
plt.title("Distribucin de sobrevivientes por cluster")
plt.show()

#Distancia entre centroides

from scipy.spatial.distance import cdist

distancias = cdist(centros, centros)
print(pd.DataFrame(distancias, columns=['C0', 'C1', 'C2'], index=['C0', 'C1', 'C2']))



```
